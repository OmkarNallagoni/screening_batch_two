**Step-9: Drop the column**

- drop is a method is used to drop the columns and rows

- **axis** argument

    - axis=0 it represnts the rows

    - axis=1 it represnts the columns

- **inplace**

    - inplace=True

        - After droping the rows or columns, the modified dataframe will be overwrite in the same name

    - inplace=False

        - Modified dataframe need to save in a new variable name

```python
numbers= [i for i in range(1,10)]
square= [i*i for i in range(1,10)]
cube=[i*i*i for i in range(1,10)]
cols=['number','square','cube']
df2=pd.DataFrame(zip(numbers,square,cube),
                columns=cols)
df2
```

```python
df2.drop('number')
# here 'number' is a column
# drop method axis=0 by defaault 
# which represnts rows
# but 'number' is  not a row
# if you want to drop the column provide axis=1
```

```python
df2.drop('number',axis=1)
# Here 'number' column will be drop
# But the modified output will not overwrite the existed dataframe
# why because the argument inplace= False by default
# change into inplace=True if you want to overwrite
```

```python
df2.drop('number',axis=1,inplace=True)
```
