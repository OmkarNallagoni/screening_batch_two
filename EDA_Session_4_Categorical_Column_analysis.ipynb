```python
# Create the frquency tables
# Import the packages
# Read the data
# cat and numerical columns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

visa_df=pd.read_csv('Visadataset.csv')

cat_cols=visa_df.select_dtypes(include='object').columns
num_cols=visa_df.select_dtypes(exclude='object').columns
```

**Bar chart**

- For categorical columns we need to represnt two ways

  - Frequency table(Tabular representation) 

  - Bar chart (Graphical represntation)

- We know that Bar charts have two axis

    - X-axis: represnts lables

    - Y-axis: represnts Count

- So inorder to plot a bar chart first we need to make a frequency table

- here we will cretate the frequency table ,and immediately will draw the bar charts

- bar plot is avaialable in **matplotlib package**

**continent column**

```python
data=visa_df['continent'].value_counts()
keys=data.keys()
values=data.values
cols=['Lables','Count']
df=pd.DataFrame(zip(keys,values),columns=cols)
```

```python
plt.figure(figsize=(10,5))
plt.xlabel('Lables')
plt.ylabel('Count')
plt.title('Bar chart')
plt.bar(df['Lables'],df['Count'])
plt.savefig('Continent_bar_chart.png')
plt.savefig('Continent_bar_chart.jpg')
plt.show()
```

**Note:plt.show always at last**

- step-1: we read the data

- step-2: we taken one cat column

- step-3: we created a frequency table

- step-4: we plot the bar chart

**Count plot**

- Count plot also same as bar chart

- Count plot under seaborn package

- Count plot does not require frequency table to plot bar chart

- Count plot need only Column name and original dataframe name

- Assume that column name: **continent**

- Original data frame name **visa_df**
